- 한 페이지 당 20개 * 총 50 페이지 = 1000개
- 리뷰 크롤링 하고 전처리 과정에서 한글 제외하는 게 좋을 것 같아서 그 코드는 제외
- 줄바꿈이 있는 리뷰가 있어서 그 부분만 replace 사용

```python
페이지 이동
버튼 1>2
#__next > div.css-edibn.e1d4aooe2 > section.e6fth966.css-1w043rb.ex98du0 > div.css-0.e13i1jpn0 > div > button

버튼 2>3 이후 동일
#__next > div.css-edibn.e1d4aooe2 > section.e6fth966.css-1w043rb.ex98du0 > div.css-0.e13i1jpn0 > div > button:nth-child(3)
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# WebDriver 초기화
driver = webdriver.Chrome()

# 페이지 로드
driver.get("https://product.29cm.co.kr/catalog/1942789?category_large_code=268100100")  # 실제 URL로 변경하세요.

# 리뷰 버튼 클릭
WebDriverWait(driver, 10).until(
    EC.element_to_be_clickable((By.XPATH, '//*[@id="__next"]/div[5]/div[2]/div[2]/div[1]/div/div[2]/button'))
).click()

reviews = []
for page_number in range(1, 51):
    try:
        # 페이지 로드 및 대기
        time.sleep(2)

        # 리뷰 텍스트 수집
        for i in range(1, 21):
            selector = f'#__next > div.css-edibn.e1d4aooe2 > section.e6fth966.css-1w043rb.ex98du0 > div.css-0.e13i1jpn0 > ul > li:nth-child({i}) > div.css-100fusk.eji1c1x2 > div > div.css-16tn2ye.eji1c1x7 > div.css-31l7gp.eji1c1x1 > p'
            try:
                review_element = WebDriverWait(driver, 10).until(
                    EC.presence_of_element_located((By.CSS_SELECTOR, selector))
                )
                review_text = review_element.text.replace('\n', '')  #줄바꿈 대체
                print(f'Page {page_number} Review {i}: {review_text}')
                reviews.append({'text': review_text})
            except Exception as e:
                print(f'Page {page_number} Review {i}: 리뷰를 찾을 수 없습니다. 오류: {e}')

        # 다음 페이지로 이동
        if page_number == 1:
            button_selector = '#__next > div.css-edibn.e1d4aooe2 > section.e6fth966.css-1w043rb.ex98du0 > div.css-0.e13i1jpn0 > div > button'
        else:
            button_selector = '#__next > div.css-edibn.e1d4aooe2 > section.e6fth966.css-1w043rb.ex98du0 > div.css-0.e13i1jpn0 > div > button:nth-child(3)'

        # 버튼이 로드될 때까지 대기
        button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.CSS_SELECTOR, button_selector))
        )
        button.click()

    except Exception as e:
        print(f'오류 발생: {e}')
        break

# 작업 완료 후 브라우저 닫기
driver.quit()

# 수집한 리뷰를 DataFrame으로 변환
df = pd.DataFrame(reviews)

# CSV 파일로 저장
df.to_csv('reviews.csv', index=False, encoding='utf-8-sig')  
```
