# 29cm 남성 상의 리뷰 크롤링

```python
# 환경 설정
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
from bs4 import BeautifulSoup
import re
import time

# 한글만 추출하는 패턴
hangul_pattern = re.compile('[가-힣]+')

# 데이터 담을 변수
review_ = []

# WebDriver 설정
driver = webdriver.Chrome()
url =

# 리뷰 수가 1000개가 넘는 상품들

# 'https://product.29cm.co.kr/catalog/1393015?category_large_code=272100100'
# 'https://product.29cm.co.kr/catalog/1450616?category_large_code=272100100'
# 'https://product.29cm.co.kr/catalog/993892?category_large_code=272100100'
# 'https://product.29cm.co.kr/catalog/1059532?category_large_code=272100100'
# 'https://product.29cm.co.kr/catalog/1428558?category_large_code=272100100'
# 'https://product.29cm.co.kr/catalog/1962858?category_large_code=272100100'
# 'https://product.29cm.co.kr/catalog/1911054?category_large_code=272100100'
# 'https://product.29cm.co.kr/catalog/1120309?category_large_code=272100100'
# 'https://product.29cm.co.kr/catalog/671590?category_large_code=272100100'
# 'https://product.29cm.co.kr/catalog/2657729?category_large_code=272100100'

driver.get(url)

# 리뷰 버튼 클릭
review_button = driver.find_element(By.CSS_SELECTOR, "#__next > div.css-uio8sw.e1d4aooe1 > div.css-1y47rmj.e18uevlq0 > div.css-3eeht.e18uevlq1 > div.css-uz7uc7.e31km200 > div > div.css-1ilqvfb.e1k6f1kl0 > button")
review_button.click()

# 다음으로 가는 버튼이 7페이지 이후로 바뀐다...7페이지 이전 이후 버튼을 따로 설정해야한다.

# 6->7 : #__next > div.css-edibn.e1d4aooe2 > section.e6fth966.css-1w043rb.ex98du0 > div.css-0.e13i1jpn0 > div > ul > li:nth-child(6) > button
# 7->8 : #__next > div.css-edibn.e1d4aooe2 > section.e6fth966.css-1w043rb.ex98du0 > div.css-0.e13i1jpn0 > div > ul > li:nth-child(7) > button
# 8->9 : #__next > div.css-edibn.e1d4aooe2 > section.e6fth966.css-1w043rb.ex98du0 > div.css-0.e13i1jpn0 > div > ul > li:nth-child(7) > button
# 9->10 : #__next > div.css-edibn.e1d4aooe2 > section.e6fth966.css-1w043rb.ex98du0 > div.css-0.e13i1jpn0 > div > ul > li:nth-child(7) > button


def collect_reviews(page_range, next_button_selector):
    for page_number in page_range:
        req = driver.page_source
        soup = BeautifulSoup(req, 'html.parser')
        
        for i in range(1, 21):
            selector = f'#__next > div.css-edibn.e1d4aooe2 > section.e6fth966.css-1w043rb.ex98du0 > div.css-0.e13i1jpn0 > ul > li:nth-child({i}) > div > div > div.css-16tn2ye.eji1c1x7 > div > p'
            review_element = soup.select_one(selector)
            if review_element:
                review_text = review_element.get_text()
                hangul_text = ''.join(hangul_pattern.findall(review_text))
                print(f'Review {i}: {hangul_text}')
                review_.append({'text': hangul_text})
        
        try:  # 다음 페이지로 이동
            next_button = driver.find_element(By.CSS_SELECTOR, next_button_selector.format(page_number + 1))
            scroll = driver.find_element(By.CSS_SELECTOR, '#__next > div.css-edibn.e1d4aooe2 > section.e6fth966.css-1w043rb.ex98du0 > div.css-0.e13i1jpn0 > ul > li:nth-child(18) > div > div > div.css-69znzl.eji1c1x4 > span.css-1riowxi.eji1c1x6')
            driver.execute_script("arguments[0].scrollIntoView();", scroll)
            time.sleep(1)
            driver.execute_script("arguments[0].click();", next_button)
            time.sleep(1)
        except Exception as e:
            print(f'Failed to click next button on page {page_number}: {e}')
            break

# 첫 번째 페이지 범위 수집
first_next_button_selector = '#__next > div.css-edibn.e1d4aooe2 > section.e6fth966.css-1w043rb.ex98du0 > div.css-0.e13i1jpn0 > div > ul > li:nth-child({}) > button'
collect_reviews(range(1, 8), first_next_button_selector)

# 두 번째 페이지 범위 수집 # 다음으로 가는 버튼 변경
second_next_button_selector = '#__next > div.css-edibn.e1d4aooe2 > section.e6fth966.css-1w043rb.ex98du0 > div.css-0.e13i1jpn0 > div > ul > li:nth-child(7) > button'
collect_reviews(range(8, 52), second_next_button_selector)
# 어째서인지 8페이지가 수집이 안된다. 원인을 모르겠어서 일단 갯수만 맞췄다.

driver.quit()
```

```python
# 데이터프레임으로 변환
df = pd.DataFrame(review_)
print(df)

# 데이터프레임을 CSV 파일로 저장
df.to_csv('re_10', index=False)
```
